```python
def encode(cols):
    for i in cols:
        traindata[i] = LabelEncoder().fit_transform(traindata[i])
        testdata[i]  = LabelEncoder().fit_transform(testdata[i])
col = ['MSZoning','LotFrontage','MSSubClass']
encode(col)
```

```python
def type(cols):
    for i in cols:
        traindata[i] = traindata[i].astype('Int64')
        testdata[i]  = testdata[i].astype('Int64')
col = ['MSZoning','LotFrontage','MSSubClass']
type(col)
```

```python
def values(cols):
    for i in cols:
        traindata[i] = traindata[i].fillna(traindata[i].median())
        testdata[i]  = testdata[i].fillna(testdata[i].median())
col = ['MSZoning','LotFrontage','MSSubClass']
values(col)
```
