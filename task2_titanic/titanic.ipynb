```python
def nullvalues():
    train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
    test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())
nullvalues()
```

```python
def age(df):
    df.loc[df['Age'] <= 20,'Age' ] = 1
    df.loc[(df['Age'] > 20) & (df['Age'] <= 40),'Age'] = 2
    df.loc[(df['Age'] > 40) & (df['Age'] <= 60),'Age'] = 3
    df.loc[(df['Age'] > 60) & (df['Age'] <= 80),'Age'] = 4
age(train_data)
age(test_data) 
```

```python
def inttype(df):
    df['Age'] = df['Age'].astype("Int64")
```

```python
inttype(train_data)
inttype(test_data)
```

```python
age(train_data)
age(test_data)
```

```python
def fill_mode(col):
    for i in col:
        train_data[col].fillna(train_data[col].mode())
        test_data[col].fillna(test_data[col].mode())
```

```python
fill_mode(['CryoSleep','VIP'])
```

```python
def type(col):
    for i in col:
        lable = LabelEncoder()
        train_data[i] = lable.fit_transform(train_data[i])
        test_data[i] = lable.fit_transform(test_data[i])
```

```python
type(['CryoSleep' , 'VIP'])
```

```python
train_data.info()
```

```python
feature = ['CryoSleep','Age','VIP']
```

```python
x = train_data[feature]
z = test_data[feature]
y = train_data.Transported
```

```python
model = RandomForestClassifier()
model.fit(x,y)
```

```python
df = pd.DataFrame(z)
df.to_csv(r'testingdata.csv',header=True , index=False)
```

```python
testingdata  = pd.read_csv(r'testingdata.csv')
```
